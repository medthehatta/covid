```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
import pandas as pd


df1 = pd.read_csv("/tmp/bMd")
df2 = pd.read_csv("/tmp/nst-est2019-popchg2010_2019.csv")
```

```python
df2_1 = df2[df2["STATE"] != 0][["NAME", "POPESTIMATE2019"]]
df3 = df2_1.set_index("NAME").join(df1.set_index("State"))
```

```python
df3["Confirmed Per Capita"] = df3["Confirmed"] / df3["POPESTIMATE2019"]
df3["Recoveries Per Capita"] = df3["Recoveries"] / df3["POPESTIMATE2019"]
df3["Deaths Per Capita"] = df3["Deaths"] / df3["POPESTIMATE2019"]
```

```python
df3[["Confirmed Per Capita", "Recoveries Per Capita", "Deaths Per Capita"]].plot(kind="bar", logy=True)
```
